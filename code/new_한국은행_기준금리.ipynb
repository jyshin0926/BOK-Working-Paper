```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.bok.or.kr/portal/singl/baseRate/list.do?dataSeCd=01&menuNo=200643'
resp = requests.get(url)
resp
```

```python
soup = BeautifulSoup(resp.content, 'lxml')
td_tag = soup.select('div table.fixed tr td')
td_tag
```

###2008년 2월 까지는 콜금리 목표, 2008년 3월부터는 한국은행 기준금리

```python
td_tag_ls = []
for td in td_tag:
    td_tag_ls.append(td.text)
td_tag_ls
```

```python
date_year_ls = []
date_day_ls = []
rate_ls = []
for i in range(0, len(td_tag_ls), 3):
    date_year_ls.append(td_tag_ls[i])
    date_day_ls.append(td_tag_ls[i+ 1])
    rate_ls.append(td_tag_ls[i+ 2]) 
```

```python
len(date_year_ls)
```

```python
import pandas as pd
df = pd.DataFrame({'Date_YY' : date_year_ls, 
                   'Date_MD' : date_day_ls, 
                   'Rate' : rate_ls})
df
```

```python
df = df.drop([44, 43, 42, 41, 40, 39, 38, 37, 36, 35, 34, 0, 1, 2, 3, 4])
```

```python
df = df.sort_values(by = 'Date_YY', ascending = True, inplace=False, ignore_index=True)
```

```python
df
```

```python
li = []
for i in range(len(df['Date_MD'])):
    li.append(df['Date_MD'][i].replace('월', '.').replace('일', '').replace(' ', ''))
li
```

```python
a = df['Date_YY'] + '.' + li
```

```python
df.insert(loc=0, column='Date', value=pd.to_datetime(a, format='%Y-%m-%d'))
```

```python
df_all = df
df_all
```

```python
del df['Date_MD'], df['Date_YY']
```
