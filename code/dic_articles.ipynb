```python
import numpy as np
training_set = np.array(articles_df[['ngram(token+ngram)','1_M_close_up_down']])
```

```python
training_set[0][0]
```

```python
from collections import defaultdict

doccnt1 = 0
doccnt0 = 0

wordfreq = defaultdict(lambda:[0,0])
pos = 0
neg = 0
neu = 0

for doc, label in training_set:
    words = doc.split(',')
    for word in words:
        if label == 'up':
            wordfreq[word][1] += 1
            pos += 1
        elif label == 'down':
            wordfreq[word][0] += 1
            neg += 1
        else:
            neu += 1
    
for key, (cnt0,cnt1) in wordfreq.items():
    if cnt1 > 0:
        doccnt0 += cnt0
    if cnt0 > 0:
        doccnt1 += cnt1

print('doccnt0 : {}'.format(doccnt0))
print('doccnt1 : {}'.format(doccnt1))
```

```python
# training : 토큰별 조건부 확률 계산

hawkish_limit = 1.3
dovish_limit = 0.76

polarity_list = []
k = 0.5
wordprobs = defaultdict(lambda : [0,0])
for key, (cnt0,cnt1) in wordfreq.items():
    wordprobs[key][0] = (cnt0 + k) / (2 * k + doccnt0)
    wordprobs[key][1] = (cnt1 + k) / (2 * k + doccnt1)
    
    neg_score = wordprobs[key][0]
    pos_score = wordprobs[key][1]
    
    polarity_score = pos_score / neg_score
    
    if polarity_score > 1.3:
        label = 1
    elif polarity_score > 0.76:
        label = -1
    else:
        label
    polarity_ = (key, cnt1, cnt0, pos_score, neg_score, polarity_score,label)
    polarity_list.append(polarity_)
```

```python
dic = pd.DataFrame(polarity_list, columns=['ngrams', 'cnt_up', 'cnt_down','pol_up','pol_down','polarity score','label'])
```

```python
dic.to_csv('polarity_dic.csv',index=False)
```
