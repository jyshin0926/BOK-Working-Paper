```
# 문장 어조 구한 것을 다시 0보다크면 'hawkish' 작으면 'dovish' 라벨 붙이고 이걸로 다시 문서 전체 tone 구하기

def tone_per_s(sentence, hawk_list, dov_list):
  hawki_cnt = 0
  dovis_cnt = 0
  mpko = MPKO(kind=1)
  tokens = mpko.tokenize(sentence) #문장별로 토큰화
  for token in tokens:
    if token in hawk_list:
      #print(token)
      hawki_cnt +=1
    if token in dov_list:
      dovis_cnt +=1
  
  if (hawki_cnt + dovis_cnt) != 0 :
    tone_s = (hawki_cnt - dovis_cnt) / (hawki_cnt + dovis_cnt)
    if tone_s > 0:
      return 1
    elif tone_s < 0:
      return -1
    else:
      return 0


# 금통위 의사록을 '.'으로 분리하고 각 문장의 어조를 tone 구하는 공식으로 구하기
def tone_per_d(p, hawk_list, dov_list):
  tmp = p.split('.')
  hawki_cnt = 0
  dovis_cnt = 0
  for sentence in tmp:
    det = tone_per_s(sentence, hawk_list, dov_list)
    if det == 1:
      hawki_cnt += 1
    elif det == -1:
      dovis_cnt += 1
    else:
      continue
  tone_d = (hawki_cnt - dovis_cnt) / (hawki_cnt + dovis_cnt)
  return tone_d
```
