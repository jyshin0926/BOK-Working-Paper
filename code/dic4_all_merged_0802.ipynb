```python
from collections import defaultdict
import re

doccnt1 = 0
doccnt0 = 0

wordfreq = defaultdict(lambda:[0,0])
pos = 0
neg = 0
neu = 0

#for doc, label in set4dic:
for i in range(len(set4dic)):
    words = set4dic[i][0].replace('.',',').split(',')
    for word in words:
        if ('특파원' or '기자' or '기사' or ' ' or "'") in word:
            words.remove(word)
        elif re.findall('[a-z]', word):
            words.remove(word)
        elif re.findall('([一-龥/A-Z]+)',word):
            words.remove(word)
            
    for word in words:
        if set4dic[i][1] == 'up':
            wordfreq[word][1] += 1
            pos += 1
        elif set4dic[i][1] == 'down':
            wordfreq[word][0] += 1
            neg += 1
        else:
            neu += 1
    

wordfreq = {key:(cnt0,cnt1) for key,(cnt0,cnt1) in wordfreq.items() if cnt0 >= 15 or cnt1 >= 15}
    
for key, (cnt0,cnt1) in wordfreq.items():
    if cnt1 >= 15:
        doccnt1 += cnt1
    if cnt0 >= 15:
        doccnt0 += cnt0

print('doccnt0 : {}'.format(doccnt0))
print('doccnt1 : {}'.format(doccnt1))
```

```python
# training : 토큰별 조건부 확률 계산

hawkish_limit = 1.3
dovish_limit = 0.76

polarity_list = []
k = 0.5
wordprobs = defaultdict(lambda : [0,0])
for key, (cnt0,cnt1) in wordfreq.items():
    wordprobs[key][0] = (cnt0 + k) / (2 * k + doccnt0)
    wordprobs[key][1] = (cnt1 + k) / (2 * k + doccnt1)
    
    neg_score = wordprobs[key][0]
    pos_score = wordprobs[key][1]
    
    polarity_score = pos_score / neg_score
    
    if polarity_score > 1.3:
        label = 1
    elif polarity_score < 0.76:
        label = -1
    elif polarity_score <= 1.3 and polarity_score >= 0.76:
        label = 0
    polarity_ = (key, cnt1, cnt0, pos_score, neg_score, polarity_score,label)
    polarity_list.append(polarity_)
```

```python
dic = pd.DataFrame(polarity_list, columns=['ngrams', 'cnt_up', 'cnt_down','w|pos','w|neg','polarity score','label'])
```
