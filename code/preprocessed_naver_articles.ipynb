### date 타입 정규화 및 통일

```python
reg_date_list = []
for i in range(len(df_articles)):
    if type(df_articles['date'][i]) != str:
        df_articles['date'][i] = str(df_articles['date'][i])
    reg_date = re.findall('[0-9.가-힣 0-9:]+',df_articles['date'][i])
    reg_date_list.append(reg_date)
for i in range(len(reg_date_list)):
    if len(reg_date_list[i]) == 1:
        reg_date_list[i] = reg_date_list[i]
    elif len(reg_date_list[i]) > 1:
        reg_date_list[i] = reg_date_list[i][-1]
    reg_date_list[i] = ''.join(reg_date_list[i])
    
df_articles['date'] = reg_date_list
```

## 기사 내용 가공

```python
def content_cleansing(content):
    tag_list = ["// flash 오류를 우회하기 위한 함수 추가","function _flash_removeCallback() {}","[","]","\'\\n\\t\'","\'\\n\\t\\n\\t","\'\\n\'","\'\\n\\n\'","\\n\\t\\n\\t","\\r","/"]
    remain_tag = ["\\n","\\n","''",",","\'","\\n\\n\\t\\t\\n\\t\\t","\\n\\n\\t\\t","\\n\\n\\t\\n\\n\\t\\n","\n\n\t","\n\t\n\t\t\n\t\t\n\t","\n\n\t\t\n\t\t","\n\n\t\t","\n\n\t\n\n\t\n"]
 
    while content:
        for x in tag_list:
            content = content.replace(x,"")
        break
    content_= re.sub("&nbsp;|\t|\r|\n", '', content)
    while content_:
        for x in remain_tag:
            content_ = content_.replace(x,"").strip()
        break
    return content_
        
def content_to_list(content):
    reg_content = []
    for i in range(len(content)):
        reg_content.append(content_cleansing(content[i]))
    return reg_content
```

```python
df_articles['content'] = content_to_list(df_articles['content'])
```
